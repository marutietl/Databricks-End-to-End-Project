### **DLT(DELTA LIVE TABLES) PIPELINE**

### **STREAMING TABLE**

```python
import dlt
from pyspark.sql.functions import *
```

```python
@dlt.table()

def DimProducts_stage():
    df=spark.readStream.table("databricks_ete_cata.silver.products_silver")

    return df
```

### **Streaming View**

```python
@dlt.view()

def DimProducts_View():
  df=spark.readStream.table("Live. DimProducts_stage")
  return df
```

### **DimProducts**

```python
dlt.create_Streaming_table("DimProducts")
```

```python
dlt.apply_changes(
  target = "DimProducts",
  source = "DimProducts_View",
  keys = ["product)id"],
  sequence_by = "product_id",
  stored_as_scd_type =2
)
```
